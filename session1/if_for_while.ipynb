# `for`

Un `for` en Python itera sobre una lista

```python
l = [4, -5.5, "hola", [3, 4]]
l
```

```python
for elt in l:
    print(elt)
```

Nota que en cada vuelta por el bucle, `elt` tiene un tipo distinto

Para imprimir todos los números de $1$ a $10$:
Necesitamos *crear* una lista de los números:

```python
print(range(10))
```

```python
for i in range(10):
    print(i)
```

podemos applicar operaciones en tot los elementos de la lista utilizando el for

```python
for i in range(10):
    print(i*10)
```

Podemos guardar esto en una nueva lista

```python
l = []
for i in range(10):
    l.append(i*10)
l
```

```python
l3=[1]
if len(l3)==0:
    print("empty list")
```

Tambien podemos applicar opeciones en todo los elementos de esta nueva lista para modificar la

```python
for i in range(10):
    l[i]=l[i]*10
l
```

es lo mismo que en syntax corta:

```python
#l=range(10)
l2=[elt/10 for elt in l]
l2
```

Podemos combinar `for` y `if`

```python
for elt in l:
    if elt%200 == 0:
        print("element %i is a multiple of 200"%elt)
    else: 
        print("element %i is not a multiple of 200"%elt)
        print("patata")
```

es lo mismo que

```python
for i in range(len(l)):
    if l[i]%200 == 0:
        print("element %i is a multiple of 200"%l[i])
    else: 
        print("element %i is not a multiple of 200"%l[i])
```

Podemos utilizar `for` para seleccionar y/o encontrar elementos

```python
indmul = []
eltmul = []
for i in range(len(l)):
    if l[i]%200==0:
        indmul.append(i)
        eltmul.append(l[i])
print(indmul, eltmul)
```

de esta manera podemos hacer operaciones sobre solo unaparte de la lista

```python
for i in indmul:
    l[i]=l[i]/2
print(l)
```

## Encontrar el maximo

Aqui os pongo unos ejemplos typicos de algoritmos sobre listas

primero generamos una lista aleatoria

```python
# import random  (aqui importamos una libreria volveremos en este concepto en la poxima classe)
from random import sample 

lista = range(0,1000)
lrand = sample(lista,k=50)
lrand
```

Encontrar el maximo de esta lista y su posicion

```python
n = len(lrand) #contamos los elementos
maxelt = lrand[0] #initializamos la variable en la cual guardaremos el maximum
pos = 0
for i in range(n):
    elt=lrand[i]
    if elt>maxelt: #comparamos el elemento con el maximo que teniamos
        maxelt=elt
        pos = i
print(maxelt, pos)
```

Mas "elegante" y rapido (pero no perite guardar la posicion facilmente)

```python
n = len(lrand) #contamos los elementos
maxelt = lrand[0] #initializamos la variable en la cual guardaremos el maximum
for i in range(n):
    maxelt = max([elt, maxelt])
print(maxelt)
```

## Construct a list of dates

Manipular fechas es essential en meterologia y climatologia, asi que tenemos que practicar bastante con esto

Puede ser que hayais escuchado hablar de friday for future, que son concentraciones, huelgas, conferencias.... para protestar contra la inaccion de los goviernos frente al cambio climatico. Tiene lugar cada viernes, y si miramos la proximas fechas de este año...

```python
#We import the date library (We will come back to the library concept later in another class)
import datetime 
date = datetime.date.today() 
listfff = []
while date.year==2019:
    listfff.append(date)
    date = date + datetime.timedelta(days=7)
```
